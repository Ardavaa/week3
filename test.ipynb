```python
import pandas as pd
import csv
from io import StringIO

with open('padi_umkm_fashion.csv', 'r', encoding='utf-8') as file:
    raw_data = file.read()

cleaned_data = raw_data.replace(',,,', '').replace(',,', '')

cleaned_data = cleaned_data.replace('"', '')

data = StringIO(cleaned_data)
reader = csv.reader(data, delimiter='|')

df = pd.DataFrame(reader)

df.columns = df.iloc[0]
df = df[1:]

df = df.reset_index(drop=True)
df = df.dropna(axis=1, how='all')

columns_to_keep = ['name', 'address', 'city_name', 'province_name', 'latitude', 'longitude', 'cust_flag']

df = df[columns_to_keep]
df
```
